```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
CONFIG = {
    'TRAIN_PATH': '/kaggle/input/datathon-entel-2022-reto2/train.csv',
    'TEST_PATH': '/kaggle/input/datathon-entel-2022-reto2/test.csv',
    'SAMPLE_SUBMISSION': '/kaggle/input/datathon-entel-2022-reto2/test_sample.csv'
}
```

```python
df_train = pd.read_csv(CONFIG['TRAIN_PATH'])
df_test = pd.read_csv(CONFIG['TEST_PATH'])
df_sub = pd.read_csv(CONFIG['SAMPLE_SUBMISSION'])
```

# Analizando por cada "z" variable

# Z_MARCA

```python
px.line(df_train.groupby('Z_MARCA')[df_train.columns[5:]].sum().reset_index(drop=True).T)
```

#### De todas las marcas, existen dos que sobresalen claramente (apple y samsung) seguramente 

```python
px.bar(df_train.groupby('Z_MARCA')[df_train.columns[5:]].sum().mean(axis=1).sort_values(ascending=False))
```

#### El promedio de venta de estás dos marcas sobrepasan los 10K por semana, mientras que los demás no llegan a 5k

```python
px.bar(df_train['Z_MARCA'].value_counts(normalize=True))
```

#### Las dos primeras marcas represetan casi el 50% de todas las ventas de celulares

# Z_GAMA

```python
px.line(df_train.groupby('Z_GAMA')[df_train.columns[5:]].sum().reset_index(drop=True).T)
```

```python
px.bar(df_train.groupby('Z_GAMA')[df_train.columns[5:]].sum().mean(axis=1).sort_values(ascending=False))
```

```python
px.bar(df_train['Z_GAMA'].value_counts(normalize=True))
```

#### Existe una gama que sobresale en ventas que las demás, el cual representa casi el 50% de los datos

# Z_MODELO

```python
px.line(df_train.groupby('Z_MODELO')[df_train.columns[5:]].sum().reset_index(drop=True).T)
```

#### A diferencia de los datos en base a la marca, los datos en base a los modelos se ven más dispersos con mayor variación.

```python
px.bar(df_train.groupby('Z_MODELO')[df_train.columns[5:]].sum().mean(axis=1).sort_values(ascending=False))
```

```python
px.bar(df_train['Z_MODELO'].value_counts(normalize=True))
```

#### El modelo que una mayor cantidad de filas (registros) solo representa el 1% del total de conjunto de datos

##### Sería una mala idea convertirlos en dummies debido a la gran cantidad de modelos que existen y la poca cantidad que se asignará a cada uno.
##### Una buena idea sería agruparlos en base a la cantidad de registros o simplemente darle un peso a cada modelo, el peso sería el porcentaje de ventas

# Z_DEPARTAMENTO

```python
px.line(df_train.groupby('Z_DEPARTAMENTO')[df_train.columns[5:]].sum().reset_index(drop=True).T)
```

#### Hay un departamento que se diferencia de los demás. Seguramente este departamento es Lima. Al esconder este departamento vemos que los demás siguen su tendencia cuando sube y baja, pero a menor escala en cantidad

```python
px.bar(df_train.groupby('Z_DEPARTAMENTO')[df_train.columns[5:]].sum().mean(axis=1).sort_values(ascending=False))
```

```python
px.bar(df_train['Z_DEPARTAMENTO'].value_counts(normalize=True))
```

# Z_PUNTO DE VENTA

```python
px.line(df_train.groupby('Z_PUNTO_VENTA')[df_train.columns[5:]].sum().reset_index(drop=True).T)
```

#### Los puntos de ventas siguen la tendencia del departamento. Existen dos que tiene un gran diferencia con los demás

```python
px.bar(df_train.groupby('Z_PUNTO_VENTA')[df_train.columns[5:]].sum().mean(axis=1).sort_values(ascending=False))
```

```python
px.bar(df_train['Z_PUNTO_VENTA'].value_counts(normalize=True))
```

#### Similar a los modelos, cada punto representa el 0.5% de los datos o menos. Realizar un dummie por cada uno sería agregar complejidad a los modelos, convertirlos en peso sería la mejor opción
